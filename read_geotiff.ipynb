John Glenn Columbus International Airport


```python
elevation_john_ch1 = gr.from_file('/content/drive/MyDrive/airport_files/USGS_13_n41w083_20200302.tif')
elevation_john_ch2 = gr.from_file('/content/drive/MyDrive/airport_files/USGS_13_n40w083_20210111.tif')
```

```python
cropped_john_ch1 = elevation_john_ch1.extract(-82.8836111119177, 40.000678, 0.075)
cropped_john_ch2 = elevation_john_ch2.extract(-82.8836111119176, 40.000556, 0.075)
```

```python
elevation_john_ch1.plot()
```

```python
elevation_john_ch2.plot()
```

```python
cropped_john_ch1.plot()
```

```python
cropped_john_ch2.plot()
```

```python
df1 = cropped_john_ch1.to_pandas()
df2 = cropped_john_ch2.to_pandas()
```

```python
df_john_ch = pd.concat([df1,df2])
```

```python
df_john_ch.head()
```

```python
from georasters import from_pandas
merged_john_ch = from_pandas(df_john_ch)
```

```python
merged_john_ch.plot()
```
